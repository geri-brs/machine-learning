```
#Import libraries
import math
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```
#Create 3 group of data around 3 center points
center_1 = np.array([1,1])
center_2 = np.array([5,5])
center_3 = np.array([8,1])

data_1 = np.random.randn(200, 2) + center_1
data_2 = np.random.randn(200,2) + center_2
data_3 = np.random.randn(200,2) + center_3

#3 group concatenate into 1
data = np.concatenate((data_1, data_2, data_3), axis = 0)

#Plot all data
plt.scatter(data[:,0], data[:,1], s=7)
```

```
# Number of clusters
k = 3
# Number of training data
n = data.shape[0]
# Number of features in the data
c = data.shape[1]

# Generate random centers, here we use sigma and mean to ensure it represent the whole data
mean = np.mean(data, axis = 0)
std = np.std(data, axis = 0)
centers = np.random.randn(k,c)*std + mean

# Plot the data and the centers generated as random
plt.scatter(data[:,0], data[:,1], s=7)
plt.scatter(centers[:,0], centers[:,1], marker='*', c='r', s=150)
plt.scatter(mean[0], mean[1], marker='+', c='g', s=150)
```

```
#This method get 2 parameters(2 points) calculates and return shortest distance between them
def Euclidean_distance(one, two):

    squared_distance = 0
    
    for i in range(len(one)):

            squared_distance += ((one[i] - two[i])**2)

    ed = math.sqrt(squared_distance)

    return ed;
  
#print(Euclidean_distance([2,0], [2,6]))
```

```
#This array numbers(0,1,2) what represents the exact group of each data point
clusters_of_data = np.zeros(shape=(600,1))

for p, point in enumerate(data):
  
  current_distance = 0
  best_distance = 100
    
  for c, center in enumerate(centers):

      current_distance = Euclidean_distance(point, center)
      if (current_distance < best_distance):
        best_distance = current_distance
        cluster_index = c
  clusters_of_data[p] = cluster_index 
  
print(clusters_of_data)
```
